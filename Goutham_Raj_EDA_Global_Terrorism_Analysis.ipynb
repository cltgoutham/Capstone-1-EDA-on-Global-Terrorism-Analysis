## 2. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
#taking only specific columns
df=df.loc[:,['iyear','imonth','iday','country_txt','region_txt','city','success',
             'attacktype1_txt','targtype1_txt','target1','natlty1_txt','gname','weaptype1_txt','nkill','nwound','propextent_txt']]

#renaming the columns
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country',
                   'region_txt':'Region','city':'City','success':'Success',
                   'attacktype1_txt':'Attacks','targtype1_txt':'Target Type','target1':'target','natlty1_txt':'nationality',
                   'gname':'Group Name','weaptype1_txt':'Weapon Type','nkill':'Killed','nwound':'Wounded','propextent_txt':'Propextent'},inplace= True)


#adding new column/concatinating column and dealing with null values
#df.isnull().sum()
df['Killed']=df['Killed'].fillna(0)
df['Wounded']=df['Wounded'].fillna(0)
df['City']=df['City'].fillna('unknown')
df['target']=df['target'].fillna('unknown')
df['Propextent']=df['Propextent'].fillna('not estimated')
df['nationality']=df['nationality'].fillna('unknown')
df.head()

```

### What all manipulations have you done and insights you found?

We cleaned the dataset and took only the required columns for visualization. All the cell with null values were cleared and made the dataset apt for processing.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### ***1.NUMBER OF TERRORIST ATTACK OVER THE YEAR FROM 1970-2017***

```python
# Chart - 2 visualization code
#code
number_of_attack=df["Year"].value_counts().rename_axis('Year').reset_index(name='Count of Terrorist Attack')

#plot- using plotly
plt.rcParams['figure.figsize'] = (10,10)
plot1=px.bar(number_of_attack, x="Year",y="Count of Terrorist Attack",title="Terrorist Attack Per Year",color='Year')
plot1.show()
```

##### 1. Why did you pick the specific chart?

Bar graph is specifically used for showing a relationship between 2 particular variables. Here we used two variables Number of attacks and Year. Hence it shows the Number of attacks on each year with a proper count. For showing the proper count on each year we use bar graph from plotly.

##### 2. What is/are the insight(s) found from the chart?

In the previous graph we saw the overall trend of terrorism over the last decades, but here we can get the proper/exact count on terrorist attacks on each year.As per the plotly bar graph, Number of terrorist attacks is much less in 1971 and is highest at 2014.

#### ***2.TERRORISM PRONE REGION IN THE WORLD***

```python
# Chart - 3 visualization code
#most affected region
affect_region=df.groupby('Region')['Attacks'].count().reset_index().sort_values(by='Attacks', ascending=False)
affect_region
```

```python
from wordcloud import WordCloud

# Create a word cloud object
wordcloud = WordCloud(width=800, height=400, background_color='black').generate_from_frequencies(affect_region)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

##### 1. Why did you pick the specific chart?

Word clouds provide a quick and easy way to visualize the most frequency of words or terms in a text or dataset. The size of each word in the cloud is typically proportional to its frequency, worcloud makes it easy to identify the most common words at a glance.

##### 2. What is/are the insight(s) found from the chart?

* Wordcloud shows the most addected region.
* Most affected region are Middle east & north africa and South asia.
* The least affected region is Australasia & Oceania.

#### ***5.TALIBAN AND THEIR MODE OF OPERATION***

```python
# Chart - 6 visualization code
#code
nott_taliban=df[df['Attacks'].isin(df['Attacks'].value_counts()[1:20].index)]
pd.crosstab(nott_taliban.Year,nott_taliban.Attacks).plot(color=sns.color_palette('magma',10))
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.show()
```

Different types of attacks executed by taliban across different years is shown in the graph. This multiple line plots in matplotlib gives us the overall trend of preffered attack types by taliban, hence we use multiple line plots here.

#### ***6.MOST USED WEAPONS FOR DISTRUCTION***

```python
weapons=df["Weapon Type"].value_counts()
weapons
```

```python
import plotly.graph_objects as go

# Create a pie chart
fig = go.Figure(data=[go.Pie(labels=weapons.index, values=weapons.values)])

# Set the layout
fig.update_layout(title="Weapon Types Distribution", title_x=0.5,
                  width=1000, height=600)

# Display the chart
fig.show()
```

##### 1. Why did you pick the specific chart?

Pie charts are effective in displaying the relative proportions of different categories within a dataset. The size of each slice in the pie represents the proportion of that category in relation to the whole. Plotly library helps to create interactive pie charts.



##### 2. What is/are the insight(s) found from the chart?

* Explosives are the most used weapon type.
* The second most used weapon is firearms. So basically they are focusing on making a loss at large extent.

Graph used here is Stacked Bar graph. When we have to plot multiple information in a single graph stacked bar is very helpful. Different values can be stacked on top of each other and can be represnted using differenet colors.

*   Increase in new category of weapons(unknown) can be seen increasing.
